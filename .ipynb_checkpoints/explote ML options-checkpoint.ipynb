```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import model_selection
from sklearn import linear_model
from sklearn import metrics
```

```python
# load the data
pheno_mat = pd.read_csv("C:/Users/bujdo/OneDrive/Dokumenty/PhD_OToole/aurora/pheno_mat_miss.csv")
bin_mat = pd.read_csv("C:/Users/bujdo/OneDrive/Dokumenty/PhD_OToole/aurora/bin_mat_miss.csv")
bags = pd.read_csv("C:/Users/bujdo/OneDrive/Dokumenty/PhD_OToole/aurora/bags.csv")
```

```python
# perpare phenomat by converting pheno to string
cols = list(bin_mat.columns)
pheno_mat = bin_mat['pheno'].values.tolist()
pheno_mat = [str(x) for x in pheno_mat]
```

```python
# prepare bin_mat by dropping some cols and converting factor to dichotomous values
bin_mat = bin_mat.drop(['pheno', 'ids'], axis=1)
def factor_to_numeric(x):
    if x =='hit': return -1
    if x =='miss':   return 1

bin_mat = bin_mat.applymap(factor_to_numeric)
bin_mat
```

```python
lm = linear_model.LogisticRegression(multi_class = 'ovr', solver = 'liblinear')
model = lm.fit(bin_mat, pheno_mat)
feature_coefs = model.coef_[0]
# print(metrics.classification_report(pheno_mat, lm.predict(bin_mat)))
```

```python
# calculate the feature importance
feature_names = bin_mat.columns
feature_importance = pd.DataFrame(feature_names, columns = ["feature"])
feature_importance["importance"] = pow(math.e, feature_coefs)
feature_importance = feature_importance.sort_values(by = ["importance"],  ascending = True)
feature_importance_plot = feature_importance.iloc[1:20,:]

from sklearn.linear_model import LogisticRegression
ax = feature_importance_plot.plot.barh(x = "feature", y = "importance")
plot.show()
```

```python
# get confusion matrix
from sklearn import metrics
predictions = model.predict(bin_mat)
cm = metrics.confusion_matrix(pheno_mat, predictions)

print(cm)
# the model is overfitted
```

```python
# look if log regression can find gfs important for L. reuteri
bin_mat = pd.read_csv("C:/Users/bujdo/OneDrive/Dokumenty/PhD_OToole/aurora/bin_mat_model.csv")
cols = list(bin_mat.columns)
pheno_mat = bin_mat['pheno'].values.tolist()
pheno_mat = [str(x) for x in pheno_mat]

# analyse only one habitat
for i in range(len(pheno_mat)):
    if pheno_mat[i] in ['1', '3', '4', '5']:
        pheno_mat[i] = '0'

bin_mat = bin_mat.drop(['pheno'], axis=1)
def factor_to_numeric(x):
    if x =='hit': return -1
    if x =='miss':   return 1

bin_mat = bin_mat.applymap(factor_to_numeric)
bin_mat
```

```python
lm = linear_model.LogisticRegression(multi_class = 'ovr', solver = 'liblinear')
model = lm.fit(bin_mat, pheno_mat)
feature_coefs = model.coef_[0]

# calculate the feature importance
feature_names = bin_mat.columns
feature_importance = pd.DataFrame(feature_names, columns = ["feature"])
feature_importance["importance"] = pow(math.e, feature_coefs)
feature_importance = feature_importance.sort_values(by = ["importance"],  ascending = True)
feature_importance_plot = feature_importance.iloc[1:20,:]

from sklearn.linear_model import LogisticRegression
ax = feature_importance_plot.plot.barh(x = "feature", y = "importance")
plot.show()
```

The importances are pretty accurate! this can be used as a proper GWAS method

```python
# try adaboost instead of random forest
import matplotlib.pyplot as plt

from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
# load the data
bin_mat = pd.read_csv("C:/Users/bujdo/OneDrive/Dokumenty/PhD_OToole/aurora/bin_mat_miss.csv")
```

```python
# perpare phenomat by converting pheno to string
cols = list(bin_mat.columns)
pheno_mat = bin_mat['pheno'].values.tolist()
pheno_mat = [str(x) for x in pheno_mat]
```

```python
# prepare bin_mat by dropping some cols and converting factor to dichotomous values
bin_mat = bin_mat.drop(['pheno', 'ids'], axis=1)
def factor_to_numeric(x):
    if x =='hit': return -1
    if x =='miss':   return 1

bin_mat = bin_mat.applymap(factor_to_numeric)
bin_mat
```

```python
bdt_real = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), n_estimators=300, learning_rate=0.3
)

bdt_discrete = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1),
    n_estimators=300,
    learning_rate=.03,
    algorithm="SAMME",
)

bdt_real.fit(bin_mat, pheno_mat)
bdt_discrete.fit(bin_mat, pheno_mat)
```

```python
from sklearn import metrics
predictions = bdt_real.predict(bin_mat)
cm = metrics.confusion_matrix(pheno_mat, predictions)

print(cm)
```

```python
from sklearn import metrics
predictions = bdt_discrete.predict(bin_mat)
cm = metrics.confusion_matrix(pheno_mat, predictions)

print(cm)
```
